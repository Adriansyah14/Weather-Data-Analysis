### Q) 11. What is the Mean value of each column againstt each 'Weather Condition'?

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

numerical_cols = data.select_dtypes(include=['number']).columns

result = data.groupby('Weather Condition')[numerical_cols].mean()

print(result)
```
